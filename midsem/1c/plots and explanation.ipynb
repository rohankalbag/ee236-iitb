# Varying Illumination Reverse Current

```python
import matplotlib.pyplot as plt
illum = [0.5e-3,1e-3,1.5e-3,2e-3,2.5e-3,3e-3]
currents = [1.109557e-02, 1.108109e-02, 1.106717e-02, 1.105376e-02, 1.104084e-02, 1.102836e-02]

plt.xlabel('Illumination in Amperes')
plt.ylabel('Reverse Current in Amperes')

plt.plot(illum, currents)
plt.show()
```

# Varying Illumination Forward Voltage

# Varying Temperature Reverse Current

```python
temp = [25,35,45,55,65,75]
currents=[1.108094e-02, 1.134879e-02, 1.161662e-02, 1.188333e-02, 1.214756e-02, 1.240627e-02]
voltages=[3.919064e-01, 3.651214e-01, 3.383377e-01, 3.116667e-01, 2.852441e-01, 2.593733e-01]

plt.xlabel('Temperature in C')
plt.ylabel('Forward Voltage in Volts')
plt.plot(temp, voltages)
plt.show()

plt.xlabel('Temperature in C')
plt.ylabel('Reverse Current in Amps')
plt.plot(temp, currents)
plt.show()
```

# Explanation for 1(c)

- All are showing linear behaviour
- Forward Voltage Decreases with increase in Temp
- Forward Voltage increase with increase in illumination

- Reverse Current Decreases with increase in illumination
- Reverse Current increases with increase in illumination

This is because the carriers increases with illumination and carriers decreases with increase in temperature
